```python
import pickle
def main():
    '''
    Objective: To write and read a list and dictionary to 
    and from a file
    Input Parameter: None
    Return Value: None
    '''
    f = open('file1', 'wb')
    pickle.dump(['hello', 'world'], f)
    pickle.dump({1:'one', 2:'two'}, f)
    f.close()
    
if __name__ == '__main__':
    main()
```

```python
f = open('file1', 'rb')
value1 = pickle.load(f)
value2 = pickle.load(f)
print(value1, value2)
f.close()
```
